## <center>CUISINE DATA ANALYSIS

#### Importing dependencies

```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

#### Importing data

```python
data=pd.read_csv('./Cuisine_rating.csv')
data
```

#### Checking basic informations and descriptions about data

```python
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.columns
```

#### Dropping un necessary columns 

```python
data.drop(columns=['User ID','Area code'],inplace=True)
data.head()
```

#### Renaming columns

```python
data.rename(columns={'Marital Status':'Marital_status','Alcohol ':'Alcohol','Food Rating':'Food_Rating','Service Rating':'Service_Rating','Overall Rating':'Overall_Rating','Often A S':'Often_A_S'},inplace=True)
data.head()
```

#### Check for null values

```python
data.isnull().mean()*100
```

#### check for duplicated items and dropping it

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

#### Creating a new  age column with respect to YOB column

```python
data['Age']=[(2022-YOB) for YOB in data['YOB']]
data.head()
```

#### Drop the YOB column

```python
data.drop(columns='YOB',inplace=True)
data.head()
```

```python
data.sample(20)
```

#### Custom function to create location vise data frame

```python
def sort_location(Location):
    data_Location=data[data.Location==Location]
    data_Location.drop(columns='Location',inplace=True)
    return data_Location
```

```python
Upper_East_Side_NY=sort_location('Upper East Side,NY')
Upper_East_Side_NY
```

## Descriptive EDA

#### 1. Top 10 restuarants with highest overall rating

```python
data.nlargest(10,'Overall_Rating')
```

#### 2. Top 3 indian cuisine restaurants in dataset	 with  highest food rating 

```python
data.query("Cuisines=='Indian'").nlargest(3,'Food_Rating')
```

#### 3. Average food  rating for each cuisines

```python
fud_rating_cuisines=data.groupby(['Cuisines'])['Food_Rating'].mean()
fud_rating_cuisines
```

#### 4. Which are the most repeting cuisines by female proffesionals in Upper East Side,NY

```python
Upper_East_Side_NY.query("Activity=='Professional' & Gender=='Female'").Cuisines.value_counts()
```

#### 5. Overall rating of each cuisines per location

```python
pivot=data.pivot_table(values='Overall_Rating',index='Location',columns='Cuisines',aggfunc='sum')
pivot
```

#### 6. Which cuisine is more popular in China Town, NY by its food rating

```python
China_Town_NY=sort_location('China Town, NY')
China_Town_NY.query("Food_Rating>=3").sort_values(by='Food_Rating',ascending=False).Cuisines.value_counts()
```

#### 7. Age category of people using more alcohol and smokes

```python
cutoff=[0,18,50,100]
labels=['Child','Adult','Old']
data['Age_type']=pd.cut(data.Age,bins=cutoff,labels=labels)
data.head()
```

```python
age_category_alco_smoke=data.query("Alcohol=='Socially' & Smoker=='Socially'")
age_category_alco_smoke.Age_type.value_counts()
```
