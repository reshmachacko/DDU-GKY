### Group by and Query

```python
#groupby
#sex vise data of survived people
sex_survived= data.groupby(['Sex'])['Survived'].sum()
sex_survived
```

```python
Pclass_survived=data.groupby(['Pclass'])['Survived'].sum()
Pclass_survived
```

```python
# 
sex_Pclass_survived=data.groupby(['Sex','Pclass'])['Survived'].sum()
sex_Pclass_survived
```

```python
Pclass__sex_survived=data.groupby(['Sex','Pclass'])['Survived'].sum()
Pclass__sex_survived
```

```python
#get the number of people who are survived by age
age_survived=data.groupby(['Survived'])['Age'].value_counts()
```

```python
age_survived
```

```python
#get the number of people in each Pclass by age
Pclass_age= data.groupby(['Pclass'])['Age'].value_counts()
Pclass_age
```

```python
Pclass_fare=data.groupby(['Pclass'])['Fare'].sum()
Pclass_fare
```

### people who are boarded from Southampton

```python
#query
boardedSouthampton= data.query("Boarded=='Southampton'")
boardedSouthampton
```

### survived people who are boarded from Southampton

```python

boardedSouthampton_survived= boardedSouthampton.query("Survived==1.0")
boardedSouthampton_survived
```
