## Graph Plotting

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data=pd.read_csv('./titanic.csv')
data.sample(20)
```

## UNIVARIATE ANALYSIS

### KDE PLOT

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=data.PassengerId)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=data.Age_wiki)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(data=data.Age)
plt.show()
```

## HISTPLOT

```python
sns.histplot(data.Fare)
plt.show()
```

```python
sns.histplot(data.Age)
plt.show()
```

## BOX PLOT


```python
sns.boxplot('Age', data=data, color='red')
plt.show()
```

```python
sns.boxplot('Fare', data=data, color='yellow')
```

## MULTIVARIATE ANALYSIS

### LINE PLOT

```python
sns.lineplot('Age','Fare',data=data, color='green')
plt.title('Age vs Fare')
plt.show()
```

```python
sns.lineplot('Age', 'Pclass', data=data)
plt.title('Age vs Pclass')
plt.show()
```

### PIE CHART

```python
data.columns
```

```python
Pclass_survived=data.groupby(['Pclass'])['Survived'].sum()
Pclass_survived
```

```python
sns.set_style('ticks')
Pclass_survived.plot.pie()
plt.legend()
plt.show()
```

```python
Sex_survived=data.groupby(['Sex'])['Survived'].sum()
Sex_survived
Sex_survived.plot.pie()
plt.legend()
plt.show()
```

```python
Sex_pclass=data.groupby(['Sex'])['Pclass'].sum()
Sex_pclass
Sex_pclass.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_sex_survived=data.groupby(['Pclass','Sex'])['Survived'].sum()
Pclass_sex_survived.reset_index()
```

```python
Pclass_sex_survived.plot.pie(autopct= '%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1),loc='upper left',borderaxespad=0)
plt.show()
```

```python
Pclass_sex_survived.plot.pie()
plt.legend(bbox_to_anchor=(3,1))
plt.show()
```

```python
Pclass_survived=data.groupby(['Pclass'])['Survived'].sum()
Pclass_survived
Pclass_survived.plot.pie(autopct='%1.3f%%')
plt.legend()
plt.show()
```

### BAR CHART

```python
plt.figure(figsize=(5,4))
sns.countplot('Sex', hue='Survived', data=data)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot('Sex', hue='Pclass', data=data)
plt.show()
```

### COUNT PLOT

```python
plt.figure(figsize=(4,3))
sns.countplot('Boarded', data=data)
plt.show()
```

```python
plt.figure(figsize=(3,4))
sns.countplot('Survived',data=data)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.set_style('darkgrid')
sns.countplot('Pclass', hue='Survived',data=data)
plt.title('Pclass:Survived vs dead')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('whitegrid')
sns.countplot('Pclass', hue='Sex',data=data, color='red')
plt.title('Pclass:Female vs male')
plt.show()
```

### FACTOR PLOT

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived',hue='Sex', data=data)
plt.show()
```

### VIOLIN PLOT

```python
sns.set_style('dark')
plt.figure(figsize=(4,3))
sns.violinplot('Sex', 'Age',hue='Survived', data=data)
plt.show()
```

### BAR PLOT

```python
plt.figure(figsize=(5,5))
sns.barplot('SibSp','Survived',data=data)
plt.title('SibSp & Survived')
plt.show()
```

### PAIR PLOT

```python
sns.pairplot(data=data)
plt.show()
```

### HEAT MAP

```python
corre=data.corr()
sns.heatmap(corre)
plt.show()
```
